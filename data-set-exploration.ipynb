```python
%pylab inline
import cv2
import matplotlib.pyplot as plt
import model
```

```python
x, y = model.load_driving_log('data/driving_log.csv')
```

```python
def print_input(i, x, y):
    fig = plt.figure()
    ax = plt.subplot()
    ax.imshow(x[i, :, :, :])
    fig.suptitle('steering {}'.format(y[i]))
```

```python
selection = y >= 0.9
# x[selection, :, :, :], y[selection]
print_input(0, x[selection, :, :, :], y[selection])
```

```python
mean = y.mean()
std = y.std()
((y - mean) / std).hist()
```
